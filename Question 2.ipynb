```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
```

```python
queensland = gpd.read_file("QLD_LGA_Feb14_LGReform_region.shp")
queensland.plot(figsize=(15,10))
```

```python
queensland = queensland.sort_values("ABBREV_NAM", ascending=True)
queensland.head(10)
```

```python
LGA = pd.read_csv("LGA codes2.csv", usecols=["SEIFA", "Population_2011"])
LGA.head(10)
```

```python
SEIFA = queensland.join(LGA)
SEIFA.head(10)
```

```python
SEIFA.plot(column="SEIFA", cmap='Greens', figsize=(23,12), k = 5, scheme = "QUANTILES", legend = True)
```

```python
SEIFA.plot(column="SEIFA", cmap='Greens', figsize=(23,12), scheme = "QUANTILES", k=5, legend = True).get_figure().savefig("map")
```
